## Use built-in functions

```python
from pyspark.sql.functions import broadcast
```

## Constants

```python
STATUS__ORDER__CANCELLED = 'Cancelled'
STATUS__ORDER_ITEM__CANCELLED = 'Cancelled'
```

# Master Datasets

```python
olist_volume_path = '/Volumes/workspace/olist_raw/olist_volume'
folder_name = 'gen_data'

# Master dataset
dataset_type = 'master_data'
```

## SCD Type 1 (master data): Status, Products, Customers

```python
status_dataset_path = f'{olist_volume_path}/{folder_name}/{dataset_type}/status/status_master_dataset.csv'
print(f'Status dataset path: {status_dataset_path}')

products_dataset_path = f'{olist_volume_path}/{folder_name}/{dataset_type}/products/'
print(f'Products path: {products_dataset_path}')

customers_dataset_path = f'{olist_volume_path}/{folder_name}/{dataset_type}/customers/'
print(f'Customers path: {customers_dataset_path}')
```

### Status (for Order and Order_Items)

```python
status_df = spark.read.csv(status_dataset_path, header=True, inferSchema=True)

# broadcast small tables
status_df = broadcast(status_df)

status_df.orderBy(status_df.type.asc(), status_df.order.asc()).show(truncate=False)
```

### Products (and built-in categories)

```python
prod_df = spark.read.csv(products_dataset_path, header=True, inferSchema=True)
prod_df.printSchema()
print('Total products : %d' % prod_df.count())
print('Total active products : %d' % prod_df.filter(prod_df.deleted_at.isNull()).count())
print('Total product categories (from Product) : %d' % prod_df.select('category').distinct().count())
```

### Customers

```python
cust_df = spark.read.csv(customers_dataset_path, header=True, inferSchema=True)
cust_df.printSchema()
print('Total customer : %d' % cust_df.count())
print('Total active customers : %d' % cust_df.filter(cust_df.deleted_at.isNull()).count())
```

# Transactional Datasets

```python
olist_volume_path = '/Volumes/workspace/olist_raw/olist_volume'
folder_name = 'gen_data'

# Master dataset
dataset_type = 'transactions_data'
```

## SCD Type 2 (transactions): Orders and Order_Items

```python
orders_dataset_path = f'{olist_volume_path}/{folder_name}/{dataset_type}/orders/'
print(f'Orders path: {orders_dataset_path}')

order_items_dataset_path = f'{olist_volume_path}/{folder_name}/{dataset_type}/order_items/'
print(f'Order Items path: {order_items_dataset_path}')
```

### Orders

```python
orders_df = spark.read.csv(orders_dataset_path, header=True, inferSchema=True)
orders_df.printSchema()
print('Total orders : %d' % orders_df.count())
print('Total active orders : %d' % orders_df.filter(orders_df.order_status != STATUS__ORDER__CANCELLED).count())
```

### Order Items

```python
order_items_df = spark.read.csv(order_items_dataset_path, header=True, inferSchema=True)
order_items_df.printSchema()
print('Total order items : %d' % order_items_df.count())
print('Total active order items : %d' % order_items_df.filter(order_items_df.item_status != STATUS__ORDER_ITEM__CANCELLED).count())
```
